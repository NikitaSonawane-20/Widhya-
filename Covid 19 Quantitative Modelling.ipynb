## Loading dataset  

```python
#Loading data from github
df = pd.read_csv('https://raw.githubusercontent.com/WidhyaOrg/datasets/master/covid19.csv')
```

```python
#Head of dataset
df.head()
```

## Data Collection and Visualization

```python
df1 = df[['Date','ConfirmedIndianNational','ConfirmedForeignNational','Cured','Deaths']]
```

```python
# Groupping by date
df1.groupby(df1.Date,sort=False).sum()
```

```python
#Create dataframe
df2 = pd.DataFrame(df['Date'])
```

```python
#Add column 'total cases' to dataframe
df2['total_cases'] = df1.sum(axis=1)
```

```python
df3 = df2.groupby(df2.Date,sort=False).sum()
```

```python
df3.tail(20)
```

```python
# Visualize data
plt.figure(figsize=(30,10))
df3.plot()
plt.show()
```
